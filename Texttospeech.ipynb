## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

You can import your own data into Colab notebooks from your Google Drive account, including from spreadsheets, as well as from Github and many other sources.
```
from gtts import gTTS
import IPython.display as ipd

# Text you want to convert to speech
text = "Hello, this is a test of text-to-speech conversion in Google Colab."

# Create a gTTS object
tts = gTTS(text)

# Save the speech as an audio file (e.g., "output.mp3")
tts.save("output.mp3")

# Play the audio file
ipd.Audio("output.mp3")
```

```
from gtts import gTTS
import IPython.display as ipd

# Ask the user for input
user_input = input("Enter the text you want to convert to speech: ")

# Create a gTTS object with the user's input
tts = gTTS(user_input)

# Save the speech as an audio file (optional)
tts.save("output.mp3")

# Play the speech
ipd.Audio("output.mp3")
```
